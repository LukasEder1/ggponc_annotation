```python
import spacy
from spacy.tokens import DocBin

def spans_to_spacy(spans):    
    nlp = spacy.blank('de')
    db = DocBin()
    
    for s in spans:
        doc = nlp(s['sentence'])
        spans = []
        for start, end, _, label in s['spans']:
            spans.append(doc.char_span(start, end, label=label))
        doc.spans['snomed'] = spans
        db.add(doc)
    
    return db
```
